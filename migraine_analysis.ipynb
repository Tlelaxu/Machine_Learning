## Training a model
The goal is to create a model and train it to be able to predict the next migraine, based on the features of the dataset


```python
from sklearn.model_selection import train_test_split
```

```python
X= mig.drop(['Intensity','Type'], axis=1)
y = mig['Intensity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Normalize/Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
#transformation
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## Neuronal creation

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()

model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))

model.add(Dense(1))
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(X_train,y_train, epochs=250)
```

```python
model.history.history
```

```python
loss_mig = pd.DataFrame(model.history.history)
```

```python
loss_mig.plot()
```

## Evaluate the model

```python
test_prediction = model.predict(X_test)
```

```python
test_prediction
```

```python
test_predict = pd.Series(test_prediction.reshape(132,))
```

```python
compare = pd.DataFrame( columns=['Prediction', 'y_test'])
```

```python
compare
```

```python
compare['Prediction'] = test_predict
```

```python
compare['y_test'] = y_test.round()
```

```python
compare
```

```python
sns.lineplot(data=compare, x='y_test', y='Prediction')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
test_prediction_round = test_prediction.round()
```

```python
print(classification_report(y_test, test_prediction_round))
```

```python
print(confusion_matrix(y_test, test_prediction_round))
```
